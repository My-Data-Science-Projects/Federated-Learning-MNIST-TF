```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt
import random
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns
import plotly.graph_objects as go
import plotly.offline as pyo

# Function to remove outliers based on Z-score
def remove_outliers(data):
    z_scores = np.abs((data - np.mean(data)) / np.std(data))
    threshold = 3
    filtered_data = data[(z_scores < threshold).all(axis=1)]
    removed_indices = np.where((z_scores >= threshold).any(axis=1))[0]
    return filtered_data, removed_indices

def load_and_preprocess_data(apply_outlier_removal=False):
    (x_train, y_train), (x_test, y_test) = mnist.load_data()
    
    if apply_outlier_removal:
        x_train, _ = remove_outliers(x_train.reshape(-1, 784))
        x_train = x_train.reshape(-1, 28, 28)
    
    x_train = x_train / 255.0
    x_test = x_test / 255.0
    
    num_clients = 10
    client_data = np.array_split(x_train, num_clients)
    client_labels = np.array_split(y_train, num_clients)
    return client_data, client_labels, x_test, y_test, num_clients

def create_model():
    model = tf.keras.Sequential([
        tf.keras.layers.Flatten(input_shape=(28, 28)),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(10, activation='softmax')
    ])
    model.compile(optimizer='sgd',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

def federated_averaging(client_models):
    new_model = create_model()
    new_weights = []
    for weights_list in zip(*[model.get_weights() for model in client_models]):
        new_weights.append(np.mean(weights_list, axis=0))
    new_model.set_weights(new_weights)
    return new_model

def train_federated_model(client_data, client_labels, num_clients, num_rounds=5, learning_rate=0.15, batch_size=256):
    global_model = create_model()
    
    for round_num in range(num_rounds):
        print(f"Round {round_num + 1}/{num_rounds}")
        client_models = []
        
        for client in range(num_clients):
            print(f"  Training on client {client + 1}/{num_clients}")
            model = create_model()
            model.set_weights(global_model.get_weights())
            
            optimizer = tf.keras.optimizers.SGD(learning_rate=learning_rate)
            for batch_start in range(0, len(client_data[client]), batch_size):
                batch_end = batch_start + batch_size
                x_batch = client_data[client][batch_start:batch_end]
                y_batch = client_labels[client][batch_start:batch_end]
                
                with tf.GradientTape() as tape:
                    predictions = model(x_batch, training=True)
                    loss = tf.keras.losses.sparse_categorical_crossentropy(y_batch, predictions)
                    loss = tf.reduce_mean(loss)
                
                grads = tape.gradient(loss, model.trainable_variables)
                optimizer.apply_gradients(zip(grads, model.trainable_variables))
            
            client_models.append(model)
        
        global_model = federated_averaging(client_models)
    
    return global_model

def clip_gradients(grads, l2_norm_clip):
    clipped_grads = []
    for grad in grads:
        norm = tf.norm(grad)
        clipped_grad = grad if norm <= l2_norm_clip else grad * l2_norm_clip / norm
        clipped_grads.append(clipped_grad)
    return clipped_grads

def add_noise(grads, noise_multiplier, l2_norm_clip, num_samples):
    noisy_grads = []
    for grad in grads:
        noise = tf.random.normal(grad.shape, stddev=noise_multiplier * l2_norm_clip / num_samples)
        noisy_grad = grad + noise
        noisy_grads.append(noisy_grad)
    return noisy_grads

def train_advanced_dp_model(client_data, client_labels, num_clients, num_rounds=5, initial_noise_multiplier=1.1, learning_rate=0.15, batch_size=256, l2_norm_clip=1.0):
    global_model = create_model()
    noise_multiplier = initial_noise_multiplier
    
    for round_num in range(num_rounds):
        print(f"Round {round_num + 1}/{num_rounds}")
        client_models = []
        
        for client in range(num_clients):
            print(f"  Training on client {client + 1}/{num_clients}")
            model = create_model()
            model.set_weights(global_model.get_weights())
            
            optimizer = tf.keras.optimizers.SGD(learning_rate=learning_rate)
            for batch_start in range(0, len(client_data[client]), batch_size):
                batch_end = batch_start + batch_size
                x_batch = client_data[client][batch_start:batch_end]
                y_batch = client_labels[client][batch_start:batch_end]
                
                with tf.GradientTape() as tape:
                    predictions = model(x_batch, training=True)
                    loss = tf.keras.losses.sparse_categorical_crossentropy(y_batch, predictions)
                    loss = tf.reduce_mean(loss)
                
                grads = tape.gradient(loss, model.trainable_variables)
                clipped_grads = clip_gradients(grads, l2_norm_clip)
                noisy_grads = add_noise(clipped_grads, noise_multiplier, l2_norm_clip, batch_size)
                optimizer.apply_gradients(zip(noisy_grads, model.trainable_variables))
            
            client_models.append(model)
        
        global_model = federated_averaging(client_models)
        
        # Update noise multiplier using adaptive strategy (e.g., geometric growth)
        noise_multiplier *= 1.1  # Example of a simple adaptive update
    
    return global_model

def plot_bounding_box(image, bbox):
    plt.imshow(image, cmap='gray')
    rect = plt.Rectangle((bbox[1], bbox[0]), bbox[3], bbox[2], linewidth=1, edgecolor='r', facecolor='none')
    plt.gca().add_patch(rect)
    plt.show()

def generate_random_bounding_box():
    top = random.randint(0, 14)
    left = random.randint(0, 14)
    height = random.randint(14, 28 - top)
    width = random.randint(14, 28 - left)
    return (top, left, height, width)

def plot_confusion_matrix_with_annotations(y_true, y_pred, title='Confusion Matrix'):
    cm = confusion_matrix(y_true, y_pred)
    tp = np.diag(cm).sum()
    fp = cm.sum(axis=0) - np.diag(cm)
    fn = cm.sum(axis=1) - np.diag(cm)
    tn = cm.sum() - (tp + fp.sum() + fn.sum())
    
    print(f"True Positives: {tp}")
    print(f"False Positives: {fp.sum()}")
    print(f"False Negatives: {fn.sum()}")
    print(f"True Negatives: {tn}")
    
    fig, ax = plt.subplots()
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax)
    ax.set_xlabel('Predicted')
    ax.set_ylabel('True')
    ax.set_title(title)
    
    # Add TP, FP, FN, TN annotations
    ax.text(0, 0, f'TP={tp}', ha='center', va='center', color='white', bbox=dict(facecolor='blue', alpha=0.5))
    ax.text(0, 1, f'FP={fp.sum()}', ha='center', va='center', color='white', bbox=dict(facecolor='blue', alpha=0.5))
    ax.text(1, 0, f'FN={fn.sum()}', ha='center', va='center', color='white', bbox=dict(facecolor='blue', alpha=0.5))
    ax.text(1, 1, f'TN={tn}', ha='center', va='center', color='white', bbox=dict(facecolor='blue', alpha=0.5))
    
    plt.show()

def plot_per_class_confusion_matrix(y_true, y_pred, title='Confusion Matrix', class_names=range(10)):
    for class_idx in class_names:
        y_true_binary = (y_true == class_idx).astype(int)
        y_pred_binary = (y_pred == class_idx).astype(int)
        cm = confusion_matrix(y_true_binary, y_pred_binary)
        
        fig, ax = plt.subplots()
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax)
        ax.set_xlabel('Predicted')
        ax.set_ylabel('True')
        ax.set_title(f'{title} for Class {class_idx}')
        plt.show()

def plot_class_confusion_matrix(y_true, y_pred, class_idx, title):
    y_true_binary = (y_true == class_idx).astype(int)
    y_pred_binary = (y_pred == class_idx).astype(int)
    cm = confusion_matrix(y_true_binary, y_pred_binary)
    
    fig, ax = plt.subplots()
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax)
    ax.set_xlabel('Predicted')
    ax.set_ylabel('True')
    ax.set_title(f'{title} for Class {class_idx}')
    plt.show()

def calculate_per_class_accuracy(y_true, y_pred, class_idx):
    y_true_binary = (y_true == class_idx).astype(int)
    y_pred_binary = (y_pred == class_idx).astype(int)
    return accuracy_score(y_true_binary, y_pred_binary)

def plot_accuracy_comparison(class_accuracies, labels, title='Accuracy Comparison'):
    fig = go.Figure()
    
    class_indices = list(range(len(class_accuracies[0])))
    
    for accuracies, label in zip(class_accuracies, labels):
        fig.add_trace(go.Scatter(
            x=class_indices,
            y=accuracies,
            mode='lines+markers',
            name=label,
            text=accuracies,
            hoverinfo='text',  # Show values on hover
        ))
    
    fig.update_layout(
        xaxis=dict(tickvals=class_indices, ticktext=[f'Class {i}' for i in class_indices]),
        xaxis_title='Classes',
        yaxis_title='Accuracy',
        title=title,
        showlegend=True,
        width=1200,
        height=800
    )
    
    pyo.iplot(fig)  # Use this for Jupyter Notebooks
    # pyo.plot(fig)  # Use this for standalone HTML file

def calculate_per_class_error(y_true, y_pred, class_idx):
    y_true_binary = (y_true == class_idx).astype(int)
    y_pred_binary = (y_pred == class_idx).astype(int)
    return 1 - accuracy_score(y_true_binary, y_pred_binary)

def plot_error_comparison(class_errors, labels, title='Error Comparison'):
    fig = go.Figure()
    
    class_indices = list(range(len(class_errors[0])))
    
    for errors, label in zip(class_errors, labels):
        fig.add_trace(go.Scatter(
            x=class_indices,
            y=errors,
            mode='lines+markers',
            name=label,
            text=errors,
            hoverinfo='text',  # Show values on hover
        ))
    
    fig.update_layout(
        xaxis=dict(tickvals=class_indices, ticktext=[f'Class {i}' for i in class_indices]),
        xaxis_title='Classes',
        yaxis_title='Error',
        title=title,
        showlegend=True,
        width=1200,
        height=800
    )
    
    pyo.iplot(fig)  # Use this for Jupyter Notebooks
    # pyo.plot(fig)  # Use this for standalone HTML file
```

```python
def main(apply_outlier_removal=False):
    client_data, client_labels, x_test, y_test, num_clients = load_and_preprocess_data(apply_outlier_removal=apply_outlier_removal)
    
    # Train Federated Learning Model
    global_model_federated = train_federated_model(client_data, client_labels, num_clients)
    
    # Evaluate Federated Learning Model
    global_model_federated.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    test_loss, test_accuracy_federated = global_model_federated.evaluate(x_test, y_test)
    print(f'Test Accuracy - Federated Learning: {test_accuracy_federated}')
    
    # Predictions for Federated Learning Model
    y_pred_federated = np.argmax(global_model_federated.predict(x_test), axis=1)
    
    # Train Advanced Differential Privacy Model
    global_model_advanced_dp = train_advanced_dp_model(client_data, client_labels, num_clients, l2_norm_clip=1.0)
    
    # Evaluate Advanced Differential Privacy Model
    global_model_advanced_dp.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    test_loss, test_accuracy_advanced_dp = global_model_advanced_dp.evaluate(x_test, y_test)
    print(f'Test Accuracy - Advanced Differential Privacy: {test_accuracy_advanced_dp}')
    
    # Predictions for Advanced Differential Privacy Model
    y_pred_advanced_dp = np.argmax(global_model_advanced_dp.predict(x_test), axis=1)
    
    # Train Differential Privacy Model
    global_model_dp = train_advanced_dp_model(client_data, client_labels, num_clients, l2_norm_clip=1.0)
    
    # Evaluate Differential Privacy Model
    global_model_dp.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    test_loss, test_accuracy_dp = global_model_dp.evaluate(x_test, y_test)
    print(f'Test Accuracy - Differential Privacy: {test_accuracy_dp}')
    
    # Predictions for Differential Privacy Model
    y_pred_dp = np.argmax(global_model_dp.predict(x_test), axis=1)
    
    # Calculate and display per-class accuracy and plot bounding box
    federated_accuracies = []
    advanced_dp_accuracies = []
    dp_accuracies = []
    
    federated_errors = []
    advanced_dp_errors = []
    dp_errors = []

    for class_idx in range(10):
        accuracy_federated = calculate_per_class_accuracy(y_test, y_pred_federated, class_idx)
        accuracy_advanced_dp = calculate_per_class_accuracy(y_test, y_pred_advanced_dp, class_idx)
        accuracy_dp = calculate_per_class_accuracy(y_test, y_pred_dp, class_idx)

        federated_accuracies.append(accuracy_federated)
        advanced_dp_accuracies.append(accuracy_advanced_dp)
        dp_accuracies.append(accuracy_dp)

        federated_errors.append(calculate_per_class_error(y_test, y_pred_federated, class_idx))
        advanced_dp_errors.append(calculate_per_class_error(y_test, y_pred_advanced_dp, class_idx))
        dp_errors.append(calculate_per_class_error(y_test, y_pred_dp, class_idx))

        print(f"Class {class_idx} - Federated Learning Accuracy: {accuracy_federated}")
        print(f"Class {class_idx} - Advanced Differential Privacy Accuracy: {accuracy_advanced_dp}")
        print(f"Class {class_idx} - Differential Privacy Accuracy: {accuracy_dp}")
        
        # Plot confusion matrices for each class
        plot_class_confusion_matrix(y_test, y_pred_federated, class_idx, title='Confusion Matrix - Federated Learning')
        plot_class_confusion_matrix(y_test, y_pred_advanced_dp, class_idx, title='Confusion Matrix - Advanced Differential Privacy')
        plot_class_confusion_matrix(y_test, y_pred_dp, class_idx, title='Confusion Matrix - Differential Privacy')
        
        # Plot bounding box for an example image of the class
        class_indices = np.where(y_test == class_idx)[0]
        if len(class_indices) > 0:
            random_index = random.choice(class_indices)
            example_image = x_test[random_index]
            example_bbox = generate_random_bounding_box()
            print(f"Bounding Box for Class {class_idx} - Federated Learning:")
            plot_bounding_box(example_image, example_bbox)
            
            # Example for Advanced Differential Privacy
            example_bbox = generate_random_bounding_box()
            print(f"Bounding Box for Class {class_idx} - Advanced Differential Privacy:")
            plot_bounding_box(example_image, example_bbox)
            
            # Example for Differential Privacy
            example_bbox = generate_random_bounding_box()
            print(f"Bounding Box for Class {class_idx} - Differential Privacy:")
            plot_bounding_box(example_image, example_bbox)
        else:
            print(f"No examples found for class {class_idx}")
    
    # Plot accuracy comparison
    class_accuracies = [federated_accuracies, advanced_dp_accuracies, dp_accuracies]
    labels = ['Federated Learning', 'Advanced Differential Privacy', 'Differential Privacy']
    plot_accuracy_comparison(class_accuracies, labels, title='Accuracy Comparison Among Models')
    
    # Plot error comparison
    class_errors = [federated_errors, advanced_dp_errors, dp_errors]
    plot_error_comparison(class_errors, labels, title='Error Comparison Among Models')

if __name__ == "__main__":
    main(apply_outlier_removal=True)
```
